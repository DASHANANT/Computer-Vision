# Mathematical Morphological Operation 

```python
import matplotlib.pyplot as plt
```

```python
import cv2
import numpy as np
im1= cv2.imread('Morphology_1.png',0)
```

```python
kernel1= np.ones((2,2),np.uint8)
kernel= np.ones((6,6),np.uint8)
```

```python
openn =   cv2.morphologyEx(im1,cv2.MORPH_OPEN,kernel1)
closedd = cv2.morphologyEx(im1,cv2.MORPH_CLOSE,kernel)

#switch the kernels
openn1 =   cv2.morphologyEx(im1,cv2.MORPH_OPEN,kernel)
closedd1 = cv2.morphologyEx(im1,cv2.MORPH_CLOSE,kernel1)


cv2.imshow('original ',im1)
cv2.imshow('openn ',openn)
cv2.imshow('closedd',closedd )
cv2.waitKey(0)
cv2.destroyAllWindows()
```
