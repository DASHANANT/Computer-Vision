# Mathematical Morphological Operation 

```python
import matplotlib.pyplot as plt
```

```python
import cv2
import numpy as np
im1= cv2.imread('Morphology_1.png',0)
```

#### Dilation- Adds pixels to Boundaries of objects in an image


```python
kernel= np.ones((2,2),np.uint8)
kernel1= np.ones((6,6),np.uint8)
```

```python
dilation = cv2.dilate(im1,kernel1,iterations =1)
erosion  = cv2.erode(im1,kernel,iterations =1)

cv2.imshow('original ',im1)
cv2.imshow('dilation ',dilation)
cv2.imshow('erode',erosion )
cv2.waitKey(0)
cv2.destroyAllWindows()
```
