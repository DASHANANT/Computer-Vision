# Histogram
A histogram is a graphical display of data using bars of different heights. 
In a histogram, each bar groups numbers into ranges.
Taller bars show that more data falls in that range. 
A histogram displays the shape and spread of continuous sample data.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageOps
```

```python
image= cv2.imread('pet.jpg')
hist=cv2.calcHist([image],[0],None,[256],[0,256])
plt.hist(image.ravel(),256,[0,256]);  # image.ravel()=Return a contiguous flattened array.
plt.show()
```

# Plotting separate histograms for each color layer

```python
c=('b','g','r')
for i ,col in enumerate(c):
    hist2=cv2.calcHist([image],[i],None,[256],[0,256])
    plt.plot(hist2,color= col)
    plt.xlim([0,256])
    plt.show()
    
```
