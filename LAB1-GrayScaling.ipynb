Grayscaling image
converting color image to shades of white and black
used in noise reduction,increses peocessing timeand simplify image

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageOps
```

```python
image=cv2.imread('pet.jpg')

cv2.imshow('original', image)
cv2.waitKey()  #by pressing any key it will show image 
cv2.destroyAllWindows()
s=Image.open('pet.jpg')
s
```

# Shortcut to save as direct gray image

```python
image1=cv2.imread('pet.jpg',0)  # 0 ,2 -stands for gray image
image2=cv2.imread('pet.jpg',3)  # else any number gives color image
cv2.imshow('original', image2)
cv2.imshow('original', image1)
cv2.waitKey()  #by pressing any key it will show image 
cv2.destroyAllWindows()
```

### converting to grayscale using cv2.COLOR_BGR2GRAY

```python
gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
cv2.imshow('grayscale',gray)
cv2.waitKey()
cv2.destroyAllWindows()
G=Image.fromarray((gray).astype(np.uint8))
G
```
