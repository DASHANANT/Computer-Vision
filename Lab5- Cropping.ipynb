# Cropping images 
it means Extracting some specific segments of that image

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

#reading image

image=cv2.imread('pet.jpg') 
image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
ori=Image.fromarray((image).astype(np.uint8))
ori
```

```python
x=image.shape[:2]
Height, Width= x[0],x[1]
```

```python
Taking random pixel value [0,1]
```

```python
x1,y1=int(Height* 0.1),int( Width* 0.1)
x2,y2=int(Height* 0.85), int(Width* 0.85)
```

```python
crop=image[x1:x2,y1:y2]
cropp=Image.fromarray((crop).astype(np.uint8))
cropp
```
