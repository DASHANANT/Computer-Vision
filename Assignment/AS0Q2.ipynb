# Converting to Gray image using Formula

```python

import matplotlib.image as mpimg

def rgb2gray(rgb):
    r, g, b = rgb[:,:,0], rgb[:,:,1], rgb[:,:,2]
    gray = 0.2989 * r + 0.5870 * g + 0.1140 * b
    return gray

img = mpimg.imread("inputAS0Q2.jpg")     
gray = rgb2gray(img)
data=gray
data =255 *( data / data.max()) #normalizes data in range 0 - 255
h = Image.fromarray((data).astype(np.uint8))
h.save("grayImgAS0Q2.png" )
h
```
