# Rotations
Using Rotation matrix function

cv2.getRotationMatrix2D(rotation_center_x,rotation_center_y,angle of rotation,scale)


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

#reading image

image=cv2.imread('pet.jpg') 
image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
print(image.shape)
x=image.shape[:2]
qH, qW= x[0],x[1]
R= cv2.getRotationMatrix2D((qW/2,qH/2),90,1)
print('Rotation Matrix-----''\n',R)
```

# Rotating anti-clockwise 90 degree

```python
img_rotate=cv2.warpAffine(image,R,(qW,qH))
rotate=Image.fromarray((img_rotate).astype(np.uint8))
rotate
```

# Rotating clockwise 90 degree

```python
R= cv2.getRotationMatrix2D((qW/2,qH/2),-90,1)
print('Rotation Matrix-----''\n',R)
img_rotate=cv2.warpAffine(image,R,(qW,qH))
rotate=Image.fromarray((img_rotate).astype(np.uint8))
rotate
```

# Rotating anti-clockwise 200 degree with 
## scaling factor =0.8

```python
R= cv2.getRotationMatrix2D((qW/2,qH/2),200,0.8)
print('Rotation Matrix-----''\n',R)
img_rotate=cv2.warpAffine(image,R,(qW,qH))
rotate=Image.fromarray((img_rotate).astype(np.uint8))
rotate
```

# Rotating clockwise 60 degree
## scaling=0.2

```python
R= cv2.getRotationMatrix2D((qW/2,qH/2),-60,0.2)
print('Rotation Matrix-----''\n',R)
img_rotate=cv2.warpAffine(image,R,(qW,qH))
rotate=Image.fromarray((img_rotate).astype(np.uint8))
rotate
```
